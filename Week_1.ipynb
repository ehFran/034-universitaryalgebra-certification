```python
# Explaining how each funtion works

# Converts strings input (including fractions) to float
def string_frac(in_string):
  '''
  Toma una cadena de entrada (in_string) que puede representar un número entero o una fracción.
  Verifica si hay una barra diagonal ("/") en la cadena para determinar si es una fracción.
  Si es una fracción, divide la cadena en numerador (n) y denominador (d), los convierte en números flotantes
  y devuelve la fracción como un número decimal.
  Si no es una fracción, simplemente convierte la cadena a un número flotante y lo devuelve.
  '''
  if "/" in in_string:
    nd = in_string.split("/")
    n = float(nd[0])
    d = float(nd[1])
    ans = n / d
    return ans

  else:
    ans = float(in_string)
    return ans

# Simplest one-step addition
def one_step_add():
  '''
  Genera aleatoriamente dos números (a y b) y muestra la expresión "x + a = b" al usuario.
  Luego, solicita al usuario ingresar el valor de x.
  Calcula la respuesta correcta (answer) restando a de b.
  Compara la respuesta del usuario con la respuesta correcta y da un mensaje de "Correcto" o "Intenta de nuevo" según corresponda.
  '''
  import random

  # Display problem
  a = random.randint(1, 10)
  b = random.randint(2, 24)
  print("x + ", a, " = ", b)
  ans = float(input("x = "))
  answer = b - a

  # Test input
  if ans == answer:
    print("Correct! \n")
  else:
    print("Try again!")
    print("The correct answer is ", answer, "\n")

# One-step addition with negative numbers
def one_step_subtract():
  '''
  Similar a one_step_add(), pero esta vez la ecuación es "a + x = b" y los valores de a y b son generados de manera que a sea un
  número negativo.
  La lógica de evaluación de la respuesta del usuario es la misma que en one_step_add().
  '''
  import random

  a = random.randint(-19, -1)
  b = random. randint(2, 24)
  print(a, " + x = ", b)
  ans = float(input("x = "))

  # Test
  answer = b - a
  if ans == answer:
    print("Correct! \n")
  else:
    print("Try again!")
    print("The correct answer is ", answer, "\n")

# One-step multiply
def one_step_mult():
  '''
  Genera aleatoriamente dos números (a y b) y muestra la expresión "a x = b" al usuario.
  Solicita al usuario ingresar el valor de x.
  Calcula la respuesta correcta (answer) dividiendo b por a.
  Utiliza la función string_frac() para convertir la entrada del usuario a un número flotante y compara con la respuesta correcta,
  proporcionando el mensaje apropiado.
  '''
  # Uses string_frac(<input string>)
  import random

  a = random.randint(1, 11)
  b = random. randint(2, 22)
  print(a, "x = ", b)
  ans_in = (input("x = "))
  answer = b/a

  # Test
  if string_frac(ans_in) == answer:
    print("Correct! \n")
  else:
    print("Try again!")
    print("The correct answer is ", answer, "\n")
```

# Linear Functions

```python
# Slope
'''
La pendiente es una medida que indica la inclinación de una línea. Más específicamente, la pendiente representa cuánto cambia
la variable dependiente (por ejemplo, y) en relación con un cambio unitario en la variable independiente (por ejemplo, x).

Para una línea recta en un plano cartesiano, la fórmula general para la pendiente (m) es:
m = Δx/Δy

​Donde:
Δy es el cambio en la variable dependiente (la diferencia entre las coordenadas y finales e iniciales).
Δx es el cambio en la variable independiente (la diferencia entre las coordenadas x finales e iniciales).

'''

x1 = 1
y1 = 7
x2 = 2
y2 = 10

slope = (y2 - y1) / (x2 - x1)

print("slope = ", slope)
```

```python
# Slope Intercept Ecuation
'''
La forma estándar de una ecuación de una línea en el plano cartesiano es la forma de pendiente-intercepto (intercepto de la ordenada).
y = mx + b

Donde:
y es la variable dependiente (ordenada en el eje y).
x es la variable independiente (abscisa en el eje x).
m es la pendiente de la línea.
b es el intercepto en el eje y, es decir, el valor de y cuando x = 0.
'''

x1 = 1
y1 = 7
x2 = 2
y2 = 10

# The slope is 'm'
m = (y2 - y1) / (x2 - x1)

# The intercept is 'b'
b = y1 - m*x1

# The full equation
print("y = ", m, "x + ", b)
```

# Common Factors & Square roots

```python
# The modulus (%) finds the reminder
print(5%3)
print(31%10)

# Use the modulus on a loop to find factors
number = 7

for test_factor in range(1, number+1):
  if number%test_factor == 0:
    print(test_factor)
```

```python
# Reduce fractions to lowest terms
numerator = 12
denominator = 24
factor = 1

# Find the common factor
for test_factor in range(1, denominator+1):
  if numerator%test_factor == 0 and denominator%test_factor == 0:
    factor = test_factor

# Divide out greatest common factor
n = int(numerator/factor)
d = int(denominator/factor)

print("Original: ", numerator, "/", denominator)
print("Reduced: ", n, "/", d)
```

```python
# Decimal to fraction

# Get decimal number to convert
digits = input("Enter a decimal number to convert: ")

# Convert to fraction
exponent = int(len(digits))-1
n = float(digits)
numerator = int(n * 10**exponent)
denominator = 10**exponent

# Reduce that fraction
factor = 1
for test_factor in range(1,denominator+1):
  if numerator%test_factor == 0 and denominator%test_factor == 0:
    factor = test_factor

# Divide out greatest common factor
num = int(numerator/factor)
den = int(denominator/factor)

# Output
print("The decimal is ", n)
print("The fraction is ", num, "/", den)
```
